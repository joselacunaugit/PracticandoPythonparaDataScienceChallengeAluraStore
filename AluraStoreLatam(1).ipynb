### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
tienda2.head()
tienda3.head()
tienda4.head()
```

#1. Análisis de facturación



```python
#print(tienda.columns)
#Suma ventas totales tienda
tienda["Precio"] = pd.to_numeric(tienda["Precio"], errors="coerce")
ventas_totales = int(tienda["Precio"].sum())

#Suma ventas totales tienda2
tienda2["Precio"] = pd.to_numeric(tienda2["Precio"], errors="coerce")
ventas_totales2 = int(tienda2["Precio"].sum())

#Suma ventas totales tienda3
tienda3["Precio"] = pd.to_numeric(tienda3["Precio"], errors="coerce")
ventas_totales3 = int(tienda3["Precio"].sum())

#Suma ventas totales tienda4
tienda4["Precio"] = pd.to_numeric(tienda4["Precio"], errors="coerce")
ventas_totales4 = int(tienda4["Precio"].sum())

print("Ventas totales tienda 1:", ventas_totales)
print("Ventas totales tienda 2:", ventas_totales2)
print("Ventas totales:tienda 3", ventas_totales3)
print("Ventas totales:tienda 4", ventas_totales4)

ventas_por_tienda = {
    "Tienda 1": ventas_totales,
    "Tienda 2": ventas_totales2,
    "Tienda 3": ventas_totales3,
    "Tienda 4": ventas_totales4
}

df_ventas = pd.DataFrame(list(ventas_por_tienda.items()), columns=["Tienda", "Facturacion"])
df_ventas_ordenado = df_ventas.sort_values(by="Facturacion", ascending=False)
print(df_ventas_ordenado)
```

# 2. Ventas por categoría

```python
ventas_por_categoria = tienda["Categoría del Producto"].value_counts()
print("Cantidad vendida de cada categoria en tienda1:")
print(ventas_por_categoria)

ventas_por_categoria2 = tienda2["Categoría del Producto"].value_counts()
print("Cantidad vendida de cada categoria en tienda2:")
print(ventas_por_categoria2)

ventas_por_categoria3 = tienda3["Categoría del Producto"].value_counts()
print("Cantidad vendida de cada categoria en tienda3:")
print(ventas_por_categoria3)

ventas_por_categoria4 = tienda4["Categoría del Producto"].value_counts()
print("Cantidad vendida de categoria producto en tienda4:")
print(ventas_por_categoria4)
```

# 3. Calificación promedio de la tienda


```python
#print(tienda.columns)
#len(tienda)
```

```python
#Calificación promedio tienda
tienda["Calificación"] = pd.to_numeric(tienda["Calificación"], errors="coerce")
suma_calificaciones = int(tienda["Calificación"].sum())
promedio_cali_tienda = suma_calificaciones / len(tienda)

#Calificación promedio tienda2
tienda2["Calificación"] = pd.to_numeric(tienda2["Calificación"], errors="coerce")
suma_calificaciones2 = int(tienda2["Calificación"].sum())
promedio_cali_tienda2 = suma_calificaciones2 / len(tienda2)

#Calificación promedio tienda3
tienda3["Calificación"] = pd.to_numeric(tienda3["Calificación"], errors="coerce")
suma_calificaciones3 = int(tienda3["Calificación"].sum())
promedio_cali_tienda3 = suma_calificaciones3 / len(tienda3)

#Calificación promedio tienda4
tienda4["Calificación"] = pd.to_numeric(tienda4["Calificación"], errors="coerce")
suma_calificaciones4 = int(tienda4["Calificación"].sum())
promedio_cali_tienda4 = suma_calificaciones4 / len(tienda4)

print("Promedio de calificaciones tienda 1:", promedio_cali_tienda)
print("Promedio de calificaciones tienda 2:", promedio_cali_tienda2)
print("Promedio de calificaciones tienda 3:", promedio_cali_tienda3)
print("Promedio de calificaciones tienda 4:", promedio_cali_tienda4)

calificaciones_por_tienda = {
    "Tienda 1": promedio_cali_tienda,
    "Tienda 2": promedio_cali_tienda2,
    "Tienda 3": promedio_cali_tienda3,
    "Tienda 4": promedio_cali_tienda4
}

df_calificaciones = pd.DataFrame(list(calificaciones_por_tienda.items()), columns=["Tienda", "Calificacion_Promedio"])
df_calificaciones_ordenado = df_calificaciones.sort_values(by="Calificacion_Promedio", ascending=False)
print(df_calificaciones_ordenado)
```

# 4. Productos más y menos vendidos

```python
ventas_por_producto = tienda["Producto"].value_counts()
print("Cantidad vendida de cada producto en tienda1:")
print(ventas_por_producto)

ventas_por_producto2 = tienda2["Producto"].value_counts()
print("Cantidad vendida de cada producto en tienda2:")
print(ventas_por_producto2)

ventas_por_producto3 = tienda3["Producto"].value_counts()
print("Cantidad vendida de cada producto en tienda3:")
print(ventas_por_producto3)

ventas_por_producto4 = tienda4["Producto"].value_counts()
print("Cantidad vendida de cada producto en tienda4:")
print(ventas_por_producto4)
```
